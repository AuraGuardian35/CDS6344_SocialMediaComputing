```python
# Deep Learning Models for Sentiment Analysis
# Person B - Social Media Computing Assignment

# Import libraries
import pandas as pd
import numpy as np
import pickle
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Deep Learning libraries
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, Dropout, Bidirectional
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.utils import to_categorical

# Transformer libraries
from transformers import pipeline, AutoTokenizer, AutoModelForSequenceClassification
import torch

# Import evaluation functions
from evaluation_utils import evaluate_model, plot_confusion_matrix

# Set random seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

print("Deep Learning Models - Person B")
print("="*50)

# Load preprocessed data
print("Loading preprocessed data from Person A...")

# Load tokenized sequences for LSTM
with open('data/tokenized_sequences.pkl', 'rb') as f:
    tokenized_data = pickle.load(f)

# Load labels
y_train = pd.read_csv('data/y_train.csv')['label_numeric'].values
y_test = pd.read_csv('data/y_test.csv')['label_numeric'].values

# Load cleaned text for transformers
df_clean = pd.read_csv('data/cleaned_data.csv')

print("Data loading complete!")
print(f"Vocabulary size: {tokenized_data['vocab_size']}")
print(f"Max sequence length: {tokenized_data['max_length']}")
print(f"Training samples: {len(y_train)}")
print(f"Testing samples: {len(y_test)}")

# Extract sequences
X_train_pad = tokenized_data['X_train_pad']
X_test_pad = tokenized_data['X_test_pad']
vocab_size = tokenized_data['vocab_size']
max_length = tokenized_data['max_length']

print(f"Training sequences shape: {X_train_pad.shape}")
print(f"Testing sequences shape: {X_test_pad.shape}")

# Convert labels to categorical for neural networks
y_train_cat = to_categorical(y_train, num_classes=3)
y_test_cat = to_categorical(y_test, num_classes=3)

print(f"Training labels shape: {y_train_cat.shape}")
print(f"Testing labels shape: {y_test_cat.shape}")

# Check class distribution
print(f"\nClass distribution in training:")
unique, counts = np.unique(y_train, return_counts=True)
for label, count in zip(unique, counts):
    sentiment = ['Negative', 'Neutral', 'Positive'][label]
    print(f"{sentiment} ({label}): {count} samples ({count/len(y_train)*100:.1f}%)")

# Model 1: Basic LSTM
print("\n" + "="*50)
print("TRAINING LSTM MODEL")
print("="*50)

# Create LSTM model
lstm_model = Sequential([
    # Embedding layer
    Embedding(
        input_dim=vocab_size,
        output_dim=100,           # Embedding dimension
        input_length=max_length,
        mask_zero=True           # Handle padding
    ),
    
    # Dropout for regularization
    Dropout(0.2),
    
    # LSTM layer
    LSTM(
        units=64,                # Hidden units
        dropout=0.2,             # Dropout for LSTM
        recurrent_dropout=0.2    # Recurrent dropout
    ),
    
    # Dense layers
    Dense(32, activation='relu'),
    Dropout(0.5),
    Dense(3, activation='softmax')  # 3 classes
])

# Compile model
lstm_model.compile(
    optimizer=Adam(learning_rate=0.001),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Display model architecture
print("LSTM Model Architecture:")
lstm_model.summary()

# Define callbacks
callbacks = [
    EarlyStopping(
        monitor='val_loss',
        patience=3,
        restore_best_weights=True,
        verbose=1
    ),
    ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.5,
        patience=2,
        min_lr=0.0001,
        verbose=1
    )
]

# Train LSTM model
print("\nTraining LSTM model...")
lstm_history = lstm_model.fit(
    X_train_pad, y_train_cat,
    batch_size=32,
    epochs=10,                   # Reasonable for student timeline
    validation_split=0.2,
    callbacks=callbacks,
    verbose=1
)

# Make predictions
print("\nMaking predictions...")
y_pred_lstm_proba = lstm_model.predict(X_test_pad)
y_pred_lstm = np.argmax(y_pred_lstm_proba, axis=1)

# Evaluate LSTM
lstm_results = evaluate_model(y_test, y_pred_lstm, "LSTM")

# Plot confusion matrix
plot_confusion_matrix(y_test, y_pred_lstm, title="LSTM - Confusion Matrix")

# Plot training history
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(lstm_history.history['accuracy'], label='Training Accuracy')
plt.plot(lstm_history.history['val_accuracy'], label='Validation Accuracy')
plt.title('LSTM Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(lstm_history.history['loss'], label='Training Loss')
plt.plot(lstm_history.history['val_loss'], label='Validation Loss')
plt.title('LSTM Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.show()

# Save LSTM model
lstm_model.save('models/deep_learning/lstm_model.h5')
print("Saved: models/deep_learning/lstm_model.h5")

# Model 2: Bidirectional LSTM (Enhanced)
print("\n" + "="*50)
print("TRAINING BIDIRECTIONAL LSTM MODEL")
print("="*50)

# Create Bidirectional LSTM model
bilstm_model = Sequential([
    # Embedding layer
    Embedding(
        input_dim=vocab_size,
        output_dim=100,
        input_length=max_length,
        mask_zero=True
    ),
    
    # Dropout
    Dropout(0.2),
    
    # Bidirectional LSTM
    Bidirectional(LSTM(
        units=64,
        dropout=0.2,
        recurrent_dropout=0.2,
        return_sequences=False    # Only return final output
    )),
    
    # Dense layers
    Dense(64, activation='relu'),
    Dropout(0.5),
    Dense(32, activation='relu'),
    Dropout(0.3),
    Dense(3, activation='softmax')
])

# Compile model
bilstm_model.compile(
    optimizer=Adam(learning_rate=0.001),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

print("Bidirectional LSTM Model Architecture:")
bilstm_model.summary()

# Train Bidirectional LSTM
print("\nTraining Bidirectional LSTM model...")
bilstm_history = bilstm_model.fit(
    X_train_pad, y_train_cat,
    batch_size=32,
    epochs=8,                    # Slightly fewer epochs
    validation_split=0.2,
    callbacks=callbacks,
    verbose=1
)

# Make predictions
y_pred_bilstm_proba = bilstm_model.predict(X_test_pad)
y_pred_bilstm = np.argmax(y_pred_bilstm_proba, axis=1)

# Evaluate Bidirectional LSTM
bilstm_results = evaluate_model(y_test, y_pred_bilstm, "Bidirectional LSTM")

# Plot confusion matrix
plot_confusion_matrix(y_test, y_pred_bilstm, title="Bidirectional LSTM - Confusion Matrix")

# Save BiLSTM model
bilstm_model.save('models/deep_learning/bilstm_model.h5')
print("Saved: models/deep_learning/bilstm_model.h5")

```

```python
# Model 3: Pre-trained BERT
print("\n" + "="*50)
print("USING PRE-TRAINED BERT MODEL (PyTorch Backend)")
print("="*50)

# Load cleaned text data for BERT
X_train_text = pd.read_csv('data/X_train.csv')['cleaned_text'].tolist()
X_test_text = pd.read_csv('data/X_test.csv')['cleaned_text'].tolist()

print("Loading pre-trained BERT model...")
print("This may take a few minutes to download...")

# Force PyTorch backend to avoid Keras 3 compatibility issues
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'  # Suppress TensorFlow warnings

# Use PyTorch-only models to avoid TensorFlow/Keras conflicts
try:
    # Try Twitter-specific RoBERTa (PyTorch version)
    bert_classifier = pipeline(
        "sentiment-analysis",
        model="cardiffnlp/twitter-roberta-base-sentiment-latest",
        framework="pt",  # Force PyTorch
        device=0 if torch.cuda.is_available() else -1
    )
    model_name = "Twitter-RoBERTa"
    print(f"Successfully loaded: {model_name}")
except Exception as e:
    print(f"Twitter RoBERTa failed: {e}")
    try:
        # Fallback to DistilBERT (lighter and more stable)
        bert_classifier = pipeline(
            "sentiment-analysis",
            model="distilbert-base-uncased-finetuned-sst-2-english",
            framework="pt",  # Force PyTorch
            device=0 if torch.cuda.is_available() else -1
        )
        model_name = "DistilBERT"
        print(f"Fallback successful: {model_name}")
    except Exception as e2:
        print(f"DistilBERT also failed: {e2}")
        try:
            # Final fallback to basic sentiment model
            bert_classifier = pipeline(
                "sentiment-analysis",
                framework="pt"  # Force PyTorch with default model
            )
            model_name = "BERT-Default"
            print(f"Using default model: {model_name}")
        except Exception as e3:
            print(f"All BERT models failed: {e3}")
            print("Skipping BERT analysis - will use only LSTM models")
            bert_classifier = None
            model_name = None

# Function to convert BERT predictions to our label format
def convert_bert_predictions(predictions, model_type="default"):
    """Convert BERT predictions to our 0,1,2 format"""
    converted = []
    
    for pred in predictions:
        label = pred['label'].upper()
        
        if model_type == "twitter-roberta":
            # Twitter RoBERTa uses LABEL_0, LABEL_1, LABEL_2
            if 'NEGATIVE' in label or 'LABEL_0' in label:
                converted.append(0)
            elif 'NEUTRAL' in label or 'LABEL_1' in label:
                converted.append(1)
            elif 'POSITIVE' in label or 'LABEL_2' in label:
                converted.append(2)
            else:
                # Default mapping based on score
                if pred['score'] < 0.4:
                    converted.append(0)
                elif pred['score'] < 0.6:
                    converted.append(1)
                else:
                    converted.append(2)
        else:
            # For DistilBERT and other standard models (POSITIVE/NEGATIVE only)
            if 'NEGATIVE' in label:
                converted.append(0)
            elif 'POSITIVE' in label:
                # Map positive to positive, but handle neutral case
                if pred['score'] > 0.8:
                    converted.append(2)  # High confidence positive
                else:
                    converted.append(1)  # Lower confidence -> neutral
            else:
                # Default case
                converted.append(1)
    
    return converted

# Only proceed with BERT if model loaded successfully
if bert_classifier is not None:
    # Make predictions with BERT (in batches to avoid memory issues)
    print("Making BERT predictions on test set...")
    batch_size = 50  # Smaller batches for stability
    bert_predictions = []

    for i in range(0, len(X_test_text), batch_size):
        batch = X_test_text[i:i+batch_size]
        try:
            batch_preds = bert_classifier(batch)
            bert_predictions.extend(batch_preds)
            
            if (i // batch_size + 1) % 10 == 0:
                print(f"Processed {i + len(batch)} / {len(X_test_text)} samples")
        except Exception as e:
            print(f"Error processing batch {i//batch_size}: {e}")
            # Create dummy predictions for failed batch
            dummy_preds = [{'label': 'NEUTRAL', 'score': 0.5} for _ in batch]
            bert_predictions.extend(dummy_preds)

    # Convert predictions
    model_type = "twitter-roberta" if "RoBERTa" in model_name else "default"
    y_pred_bert = convert_bert_predictions(bert_predictions, model_type)

    print(f"BERT predictions completed!")
    print(f"Prediction distribution: {np.bincount(y_pred_bert)}")

    # Evaluate BERT
    bert_results = evaluate_model(y_test, y_pred_bert, f"{model_name}")

    # Plot confusion matrix
    plot_confusion_matrix(y_test, y_pred_bert, title=f"{model_name} - Confusion Matrix")

    # Save BERT predictions
    bert_pred_data = {
        'predictions': y_pred_bert,
        'raw_predictions': bert_predictions,
        'model_name': model_name
    }

    with open('models/transformers/bert_predictions.pkl', 'wb') as f:
        pickle.dump(bert_pred_data, f)

    print("Saved: models/transformers/bert_predictions.pkl")

else:
    print("BERT model not available - skipping transformer analysis")
    bert_results = None
    y_pred_bert = None

# Model Comparison with Traditional ML
print("\n" + "="*60)
print("COMPARISON WITH TRADITIONAL ML MODELS")
print("="*60)

# Load Person A's results
try:
    with open('models/traditional_ml/results_summary.pkl', 'rb') as f:
        traditional_results = pickle.load(f)
    
    print("Loaded traditional ML results from Person A")
    
    # Combine all results
    deep_learning_models = [
        {
            'Model': 'LSTM',
            'Accuracy': lstm_results['accuracy'],
            'F1-Score': lstm_results['f1_weighted'],
            'Precision': lstm_results['precision'],
            'Recall': lstm_results['recall']
        },
        {
            'Model': 'Bidirectional LSTM', 
            'Accuracy': bilstm_results['accuracy'],
            'F1-Score': bilstm_results['f1_weighted'],
            'Precision': bilstm_results['precision'],
            'Recall': bilstm_results['recall']
        }
    ]
    
    # Add BERT results if available
    if bert_results is not None:
        deep_learning_models.append({
            'Model': model_name,
            'Accuracy': bert_results['accuracy'], 
            'F1-Score': bert_results['f1_weighted'],
            'Precision': bert_results['precision'],
            'Recall': bert_results['recall']
        })
    
    all_results = traditional_results['all_results'] + deep_learning_models
    
except:
    print("Could not load traditional ML results, showing only deep learning results")
    deep_learning_models = [
        {
            'Model': 'LSTM',
            'Accuracy': lstm_results['accuracy'],
            'F1-Score': lstm_results['f1_weighted'],
            'Precision': lstm_results['precision'],
            'Recall': lstm_results['recall']
        },
        {
            'Model': 'Bidirectional LSTM',
            'Accuracy': bilstm_results['accuracy'],
            'F1-Score': bilstm_results['f1_weighted'],
            'Precision': bilstm_results['precision'],
            'Recall': bilstm_results['recall']
        }
    ]
    
    # Add BERT results if available
    if bert_results is not None:
        deep_learning_models.append({
            'Model': model_name,
            'Accuracy': bert_results['accuracy'],
            'F1-Score': bert_results['f1_weighted'],
            'Precision': bert_results['precision'],
            'Recall': bert_results['recall']
        })
    
    all_results = deep_learning_models

# Create comparison DataFrame
comparison_df = pd.DataFrame(all_results)
print("\nComplete Model Comparison:")
print(comparison_df.round(4))

# Find best overall model
best_model_idx = comparison_df['F1-Score'].idxmax()
best_model = comparison_df.iloc[best_model_idx]['Model']
best_f1 = comparison_df.iloc[best_model_idx]['F1-Score']
print(f"\nBest overall model: {best_model} (F1-Score: {best_f1:.4f})")

# Visualize complete comparison
plt.figure(figsize=(15, 10))

# F1-Score comparison
plt.subplot(2, 2, 1)
colors = ['skyblue' if 'LSTM' in model or 'BERT' in model or 'RoBERTa' in model 
          else 'lightgreen' for model in comparison_df['Model']]
bars = plt.bar(comparison_df['Model'], comparison_df['F1-Score'], color=colors, alpha=0.8)
plt.title('All Models F1-Score Comparison')
plt.ylabel('F1-Score')
plt.xticks(rotation=45)
for i, v in enumerate(comparison_df['F1-Score']):
    plt.text(i, v + 0.01, f'{v:.3f}', ha='center')

# Accuracy comparison
plt.subplot(2, 2, 2)
plt.bar(comparison_df['Model'], comparison_df['Accuracy'], color=colors, alpha=0.8)
plt.title('All Models Accuracy Comparison')
plt.ylabel('Accuracy')
plt.xticks(rotation=45)
for i, v in enumerate(comparison_df['Accuracy']):
    plt.text(i, v + 0.01, f'{v:.3f}', ha='center')

# Precision vs Recall scatter
plt.subplot(2, 2, 3)
for i, model in enumerate(comparison_df['Model']):
    color = 'blue' if 'LSTM' in model or 'BERT' in model or 'RoBERTa' in model else 'green'
    plt.scatter(comparison_df.iloc[i]['Precision'], comparison_df.iloc[i]['Recall'], 
               color=color, s=100, alpha=0.7)
    plt.annotate(model, (comparison_df.iloc[i]['Precision'], comparison_df.iloc[i]['Recall']),
                xytext=(5, 5), textcoords='offset points', fontsize=8)
plt.xlabel('Precision')
plt.ylabel('Recall')
plt.title('Precision vs Recall')
plt.grid(True, alpha=0.3)

# Model type performance
plt.subplot(2, 2, 4)
traditional_models = ['Logistic Regression', 'Random Forest', 'Support Vector Machine']
deep_models = ['LSTM', 'Bidirectional LSTM']
transformer_models = [m for m in comparison_df['Model'] if 'BERT' in m or 'RoBERTa' in m]

traditional_f1 = comparison_df[comparison_df['Model'].isin(traditional_models)]['F1-Score'].mean()
deep_f1 = comparison_df[comparison_df['Model'].isin(deep_models)]['F1-Score'].mean()
transformer_f1 = comparison_df[comparison_df['Model'].isin(transformer_models)]['F1-Score'].mean()

categories = ['Traditional ML', 'Deep Learning', 'Transformers']
scores = [traditional_f1, deep_f1, transformer_f1]
plt.bar(categories, scores, color=['lightgreen', 'skyblue', 'orange'], alpha=0.8)
plt.title('Model Category Performance')
plt.ylabel('Average F1-Score')
for i, v in enumerate(scores):
    plt.text(i, v + 0.01, f'{v:.3f}', ha='center')

plt.tight_layout()
plt.show()

# Save complete comparison
comparison_df.to_csv('models/deep_learning/complete_model_comparison.csv', index=False)
print("\nSaved: models/deep_learning/complete_model_comparison.csv")

# Example predictions
print("\n" + "="*60)
print("EXAMPLE PREDICTIONS")
print("="*60)

# Example texts
example_texts = [
    "The flight was amazing and the crew was very helpful!",
    "Terrible service, delayed flight, worst experience ever.", 
    "The flight was okay, nothing special but not bad either.",
    "Love flying with this airline, always on time!",
    "Can't believe how rude the staff was today."
]

print("Comparing predictions from different model types:")
print("-" * 60)

# Simple preprocessing for examples
def simple_preprocess(text):
    import re
    text = text.lower()
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

# Load tokenizer for LSTM predictions
with open('data/tokenizer.pkl', 'rb') as f:
    tokenizer = pickle.load(f)

sentiment_labels = ['Negative', 'Neutral', 'Positive']

for i, text in enumerate(example_texts):
    print(f"\nExample {i+1}: '{text}'")
    print("-" * 40)
    
    # Preprocess text
    clean_text = simple_preprocess(text)
    
    # LSTM prediction
    text_seq = tokenizer.texts_to_sequences([clean_text])
    text_pad = tf.keras.preprocessing.sequence.pad_sequences(text_seq, maxlen=max_length)
    lstm_pred_proba = lstm_model.predict(text_pad, verbose=0)[0]
    lstm_pred = np.argmax(lstm_pred_proba)
    
    # BiLSTM prediction
    bilstm_pred_proba = bilstm_model.predict(text_pad, verbose=0)[0]
    bilstm_pred = np.argmax(bilstm_pred_proba)
    
    # BERT prediction (only if available)
    if bert_classifier is not None:
        bert_pred_raw = bert_classifier(text)[0]
        model_type = "twitter-roberta" if "RoBERTa" in model_name else "default"
        bert_pred = convert_bert_predictions([bert_pred_raw], model_type)[0]
        
        print(f"LSTM: {sentiment_labels[lstm_pred]} (confidence: {lstm_pred_proba[lstm_pred]:.3f})")
        print(f"BiLSTM: {sentiment_labels[bilstm_pred]} (confidence: {bilstm_pred_proba[bilstm_pred]:.3f})")
        print(f"{model_name}: {sentiment_labels[bert_pred]} (confidence: {bert_pred_raw['score']:.3f})")
    else:
        print(f"LSTM: {sentiment_labels[lstm_pred]} (confidence: {lstm_pred_proba[lstm_pred]:.3f})")
        print(f"BiLSTM: {sentiment_labels[bilstm_pred]} (confidence: {bilstm_pred_proba[bilstm_pred]:.3f})")
        print(f"BERT: Not available")

# Save final results
deep_learning_results = {
    'lstm_results': lstm_results,
    'bilstm_results': bilstm_results,
    'bert_results': bert_results,
    'best_model': best_model,
    'best_f1_score': best_f1,
    'complete_comparison': comparison_df.to_dict('records')
}

with open('models/deep_learning/deep_learning_results.pkl', 'wb') as f:
    pickle.dump(deep_learning_results, f)

print("\n" + "="*60)
print("DEEP LEARNING MODELS COMPLETE!")
print("="*60)

print(f"\nSummary:")
print(f"- Trained LSTM and Bidirectional LSTM models")
print(f"- Used pre-trained {model_name} for comparison")
print(f"- Best deep learning model: {best_model} (F1-Score: {best_f1:.4f})")
print(f"- All models saved in models/deep_learning/ and models/transformers/")
print(f"- Ready for ABSA implementation and dashboard creation")

print("\nFiles created:")
print("✓ models/deep_learning/lstm_model.h5")
print("✓ models/deep_learning/bilstm_model.h5") 
print("✓ models/deep_learning/complete_model_comparison.csv")
print("✓ models/deep_learning/deep_learning_results.pkl")
print("✓ models/transformers/bert_predictions.pkl")

print("\n🎉 Person B's deep learning work complete!")
print("Ready to collaborate with Person A on ABSA and dashboard!")
```
